```python
import pandas as pd

file_path = "C:\\Users\\mirth_gd7dvu\\Downloads\\archive\\sales.csv"
data = pd.read_csv(file_path)

# Check the first few rows and dataset info
print(data.head())
print(data.info())
```

```python
import matplotlib.pyplot as plt

# Example for a categorical column
categorical_column = 'product_category'
data[categorical_column].value_counts().plot(kind='bar', figsize=(10, 6))
plt.title(f'Distribution of {categorical_column}')
plt.xlabel(categorical_column)
plt.ylabel('Frequency')
plt.show()
```

```python
#DATA CLEANING
# Check for missing values
print(data.isnull().sum())

# Fill missing values for numerical columns with the mean
data['Sales'] = data['Sales'].fillna(data['Sales'].mean())

# Fill missing values for categorical columns with the mode
data['Category'] = data['Category'].fillna(data['Category'].mode()[0])

# Drop rows with significant missing data if needed
data = data.dropna()
```
